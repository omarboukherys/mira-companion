```python
from langchain_core.messages import HumanMessage
from mira.graph.graph import graph
```

```python
def test_routing(user_message: str):
    """Send a message, show branch fired + the reply."""
    result=graph.invoke({
        "messages":[HumanMessage(content=user_message)],
        "workflow":""
    })

    print(f"USER: {user_message}")
    print(f"WORKFLOW: {result['workflow']}")
    print(f"MIRA: {result['messages'][-1].content}")
    print("-"*70)
```
