```python
from langchain_core.messages import HumanMessage
from mira.graph.graph import get_compiled_graph
```

```python
async def chat(thread_id: str, user_message: str):
    """Send a message through the graph with persistent memory."""

    async with get_compiled_graph() as graph:
        result=await graph.ainvoke(
            {"messages":[HumanMessage(content=user_message)]},
            config={"configurable":{"thread_id": thread_id}},
        )

    last_reply=result['messages'][-1]
    print(f"USER ({thread_id}): {user_message}")
    print(f"MIRA: {last_reply.content}")
    print(f"Messages in state: {len(result['messages'])}")
    print("-"*70)
```

```python
await chat(
    thread_id="omar-test-1",
    user_message="Hey Mira! My name is Omar and I love astrobiology too.",
)
```
