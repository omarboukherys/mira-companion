```python
from langchain_core.messages import HumanMessage
from mira.graph.utils.chains import get_character_response_chain
```

```python
chain=get_character_response_chain()
print("Ready !")
```

```python
response=chain.invoke(
    {
        "messages": [HumanMessage(content="Hey! who are you ?")]
    }
)
```

```python
print(response.content)
```

```python
from langchain_core.messages import AIMessage

response = chain.invoke({
    "messages": [
        HumanMessage(content="Hey Mira!"),
        AIMessage(content="Salam! How's your day going?"),
        HumanMessage(content="Quick question — what's your favorite ML topic?"),
    ]
})

print(response.content)
```
