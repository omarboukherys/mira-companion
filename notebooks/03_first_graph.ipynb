```python
from langchain_core.messages import HumanMessage
from mira.graph.graph import graph
```

```python
result=graph.invoke({
    "messages": [HumanMessage(content="Hey Mira! what's up ?")]
})
```

```python
print("Final messages count: ", len(result['messages']))
```

```python
print("Mira: ", result['messages'][-1].content)
```

```python
from langchain_core.messages import AIMessage

result = graph.invoke({
    "messages": [
        HumanMessage(content="Hey Mira!"),
        AIMessage(content="Salam! How can I help?"),
        HumanMessage(content="What's your favorite ML topic right now?"),
    ]
})

print("Final messages count:", len(result["messages"]))
print()
print("Mira:", result["messages"][-1].content)
```
