```python
from langchain_core.messages import HumanMessage
from mira.graph.graph import get_compiled_graph


async def chat(thread_id: str, user_id: str, user_message: str):
    """Send a message through the graph with both memory layers active."""
    async with get_compiled_graph() as graph:
        result = await graph.ainvoke(
            {
                "messages": [HumanMessage(content=user_message)],
                "user_id": user_id,
            },
            config={"configurable": {"thread_id": thread_id}},
        )

    print(f"USER [{thread_id} / {user_id}]: {user_message}")
    print(f"MIRA: {result['messages'][-1].content}")
    print("-" * 70)
```

```python
await chat(
    thread_id="thread-A",
    user_id="omar-real",
    user_message="Hey Mira! Fun fact: I have a cat named Simba and I'm learning Japanese.",
)
```
