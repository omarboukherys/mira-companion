```python
from langchain_groq import ChatGroq
from mira.settings import settings
```

```python
llm=ChatGroq(
    model=settings.TEXT_MODEL_NAME,
    api_key=settings.GROQ_API_KEY,
    temperature=0.7
)
```

```python
from langchain_core.messages import HumanMessage
```

```python
response=llm.invoke(
    [HumanMessage(content="Hello! can you introduce yourself ?")]
)
```

```python
print(response.content)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
```

```python
messages = [
    SystemMessage(content="You are a pirate. Always respond in pirate slang. Keep responses short."),
    HumanMessage(content="What's the weather like today?"),
    AIMessage(content="Arrr, the sky be cloudy with a chance o' rain, matey!"),
    HumanMessage(content="And tomorrow?"),
]
```

```python
response=llm.invoke(messages)
```

```python
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import (HumanMessage, AIMessage, SystemMessage)
```

```python
prompt=ChatPromptTemplate.from_messages(
    [
        ("system", "You are helpful assistant. Be brief."),
        MessagesPlaceholder("messages")
    ]
)
```

```python
rendered=prompt.invoke(
    {"messages": [HumanMessage(content="Hello"),
                  AIMessage(content="I am robot"),
                  HumanMessage(content="What can you do ?")]}
)
```

```python
for msg in rendered.messages:
    print(f"=== {type(msg).__name__} ===")
    print(msg.content)
    print()
```
